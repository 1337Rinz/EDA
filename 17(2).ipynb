```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataLink = "https://raw.githubusercontent.com/1337Rinz/DATA_for_machine_learning/main/IRIS.csv"
```

```python
df = pd.read_csv(dataLink)
df.head()
```

```python
df.info()
```

```python
# check null values

df.isnull().sum()
```

```python
# remove NaN row

df.dropna(how = 'all', inplace = True)
df.head()
```

```python
df.shape
```

150 row / 5 features

```python
# check duplicate

df.duplicated().sum()
```

```python
# option (maybe remove if necessary)

df = df.drop_duplicates()
df.shape
```

```python
# some info of data field

df.describe()
```

```python
# visualization species type amount

sns.countplot(data = df, x = 'species')
plt.xlabel('species')
plt.ylabel('count')
plt.title('amount')
plt.show()
```

```python
# colors map for species

species_colors = {'setosa': 'blue', 'versicolor': 'orange', 'virginica': 'green'}
```

```python
# Distribution of data


# Get the numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(20,15))

# Flatten the axes array to iterate over it
axs = axs.flatten()

# Plot each histogram in its own subplot
for ax, column in zip(axs, numeric_columns):
    sns.histplot(data=df, x=column, hue='species', palette=species_colors, bins=30, ax=ax)

# Remove the unused subplot (if any)
if len(numeric_columns) < len(axs):
    for i in range(len(numeric_columns), len(axs)):
        fig.delaxes(axs[i])

plt.tight_layout()
plt.show()
```

```python
# heatmap of correaltion

plt.figure(figsize=(10,7))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# correlation between the features

sns.pairplot(df, hue="species")
plt.show()
```

```python
# swarnmplot sepal_length


sns.swarmplot(x='species', y='sepal_length', data=df, palette=species_colors.values())
plt.show()
```

```python
# swarnmplot sepal_width

sns.swarmplot(x='species', y='sepal_width', data=df, palette=species_colors.values())
plt.show()
```

```python
# swarnmplot petal_length

sns.swarmplot(x='species', y='petal_length', data=df, palette=species_colors.values())
plt.show()
```

```python
# swarnmplot petal_width

sns.swarmplot(x='species', y='petal_width', data=df, palette=species_colors.values())
plt.show()
```

```python
# Correalation map

# Perform one-hot encoding on the 'species' column
df_encoded = pd.get_dummies(df, columns=['species'])

# Compute the correlation matrix for the whole DataFrame
correlation_matrix = df_encoded.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

# data Temperature Readings : IOT Devices

```python
datalink2 = "https://raw.githubusercontent.com/1337Rinz/DATA_for_machine_learning/main/IOT-temp.csv"
```

```python
data2 =  pd.read_csv(datalink2)
data2.head()
```

```python
# data shape

data2.shape
```

```python
# Check null

data2.isnull().sum()
```

```python
data2.info()
```

```python
data2.describe()
```

max temp = 51  /   min tempt = 21

```python
# check how many  values in room_id feature.

data2['room_id/id'].nunique()
```

```python
# its 1 so maybe we dont need that features now :3 . say good bye to 'room_id/id' :(

data2.drop('room_id/id', axis=1, inplace=True)
data2
```

```python
# Visualize distribution of "temp"

data2['temp'].hist(bins=30)
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Temperature Distribution')
plt.show()
```

```python
# check frequency of 'out/in'   (we can see that data not balance... you can do something later ;) )

data2['out/in'].value_counts().plot(kind='bar')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.title('Temperature Reading Locations')
plt.show()
```

```python
# convert noted_date -> datetime format and create some new features

data2['noted_date'] = pd.to_datetime(data2['noted_date'])
data2['hour'] = data2['noted_date'].dt.hour
data2['weekday'] = data2['noted_date'].dt.weekday
```

```python
# check relationship of temperature when out/in

sns.boxplot(x='out/in', y='temp', data=data2)
plt.title('Temperature Inside vs. Outside')
plt.show()
```

```python
# check the relationship temp - time

data2.groupby('hour')['temp'].mean().plot()
plt.ylabel('Average Temperature')
plt.title('Temperature by Hour of the Day')
plt.show()

data2.groupby('weekday')['temp'].mean().plot()
plt.ylabel('Average Temperature')
plt.title('Temperature by Day of the Week')
plt.show()
```

```python
# overrall temperature pattern over time

data2.set_index('noted_date').resample('D')['temp'].mean().plot()
plt.ylabel('Average Temperature')
plt.title('Temperature Trends Over Time')
plt.show()
```
